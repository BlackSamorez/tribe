```python
%env CUDA_VISIBLE_DEVICES=8

import torch
```

```python
MODEL="Llama-3.1-8B"
BITS=2
DECODE_MODE="3inst_fp4"
GROUP_SIZE="16"
HADAMARD_SIZE="128"
EXTRA_WSCALING_SCHEME="nvfp4"
EXTRA_APPENDIX = f"-w{EXTRA_WSCALING_SCHEME}" if EXTRA_WSCALING_SCHEME is not None else ""

TORCH_CKPT=f"/nfs/scistore19/alistgrp/apanfero/models/QTIP/{MODEL}-{BITS}bit-{DECODE_MODE}-gs{GROUP_SIZE}-hs{HADAMARD_SIZE}{EXTRA_APPENDIX}"

nvfp4_errors = {}

for layer_idx in range(32):
    for layer_name in ["q", "k", "v", "o", "up", "gate", "down"]:
        data = torch.load(TORCH_CKPT + f"/{layer_idx}_{layer_name}.pt", map_location="cpu")
        nvfp4_errors[(layer_idx, layer_name)] = data['proxy_err']
```

```python
MODEL="Llama-3.1-8B"
BITS=2
DECODE_MODE="3inst_fp4"
GROUP_SIZE="16"
HADAMARD_SIZE="128"
EXTRA_WSCALING_SCHEME=None
EXTRA_APPENDIX = f"-w{EXTRA_WSCALING_SCHEME}" if EXTRA_WSCALING_SCHEME is not None else ""

TORCH_CKPT=f"/nfs/scistore19/alistgrp/apanfero/models/QTIP/{MODEL}-{BITS}bit-{DECODE_MODE}-gs{GROUP_SIZE}-hs{HADAMARD_SIZE}{EXTRA_APPENDIX}"

fp4_errors = {}

for layer_idx in range(32):
    for layer_name in ["q", "k", "v", "o", "up", "gate", "down"]:
        data = torch.load(TORCH_CKPT + f"/{layer_idx}_{layer_name}.pt", map_location="cpu")
        fp4_errors[(layer_idx, layer_name)] = data['proxy_err']
```

```python
for key, nv_err in nvfp4_errors.items():
    base_err = fp4_errors[key]
    print(f"{key}: {nv_err / base_err:.3f}")
```
